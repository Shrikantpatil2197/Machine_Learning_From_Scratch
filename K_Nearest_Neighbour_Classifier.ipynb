#### Q3_a) Write a program to find the k-nearest-neighbor density for n (unordered) points in one dimension. Use your program to plot such a density estimate for the x1 values in category ω3 in the table above for k = 1, 3 and 5.

**1. k-nearest-neighbor density for n (unordered) points in one dimension** 

```python
np.random.seed(42)
n = np.random.rand(100)
plt.hist(n)
plt.show()
```

```python
plt.scatter(np.arange(0,100),n)
plt.show()
```

```python
def k_density(point,x_train,k):
    count = 0
    h = np.arange(0,np.max(x_train*2),0.00001)
    x_neighbours = []
    for h_value in h:
        for x in x_train:
            if (abs(x-point) <= h_value/2 and x!=point and x not in x_neighbours):
                count = count + 1
                x_neighbours.append(x)
        if count<k:
            continue
        else:
            break  
    return count,x_neighbours,h_value
```

```python
#Let 
point = 0.6; k=1
count,x_neighbours,h_value = k_density(point,n,k)
print(f"h value at given point for k = {k} is :",h_value)
print('no. of neighbours in final h estimate:',count)
print('neighbours are:',x_neighbours)
```

#### Q3_b) Write a program to find the k-nearest-neighbor density estimate for n points in two dimensions.Use your program to plot such a density estimate for the (x1, x2)^t values in ω2 for k = 1, 3 and 5.

```python
def k_density_nd(point,x_train,k):
    count = 0
    h = np.arange(0,np.max(x_train*2),0.0001)
    x_neighbours = []
    for h_value in h:
        for i in range(x_train.shape[1]):
            if ((abs(x_train[:,i]-point) <= h_value/2).all() and (x_train[:,i]!=point).all() and x_train[:,i] not in np.array(x_neighbours)):
                count = count + 1
                x_neighbours.append(x_train[:,i])
        if count<k:
            continue
        else:
            break  
    return count,x_neighbours,h_value
```

```python
# (x1, x2)^t values in ω2 are
X_w2 = np.stack((x1_w2,x2_w2),axis=0)
print(X_w2,'\n')
plt.scatter(x1_w2,x2_w2)
plt.show()
for k in [1,3,5]:
    print(f'for k = {k}:')
    points = []
    window_h = []
    xneighbours = []
    counts = []
    for i in range(X_w2.shape[1]):
        point = X_w2[:,i]
        points.append(point)
        count,neighbours,h_value= k_density_nd(point,X_w2,k)
        xneighbours.append(neighbours)
        window_h.append(h_value)
        counts.append(count)
    data = pd.DataFrame({'points':points,'neighbour points':xneighbours,
                            'windowheight(h)':window_h,'count':counts})
    display(data)
    print('\n')
```

#### Q3_C) Write a program to form a k-nearest-neighbor classifier for the three-dimensional data from the three categories in the table above. Use your program with k = 1, 3 and 5 to estimate the relative densities at the following points: (−0.41, 0.82, 0.88)^t, (0.14, 0.72, 4.1)^t and (−0.81, 0.61,−0.38)^t.

```python
def k_classifier(neighbours,X_W1,X_W2,X_W3):
    count_w1, count_w2, count_w3 = 0, 0, 0
    for array in neighbours:
        for i in range(X_W1.shape[1]):
            if (array == X_W1[:,i]).all():
                count_w1 +=1
            elif (array == X_W2[:,i]).all():
                count_w2 +=1
            elif (array == X_W3[:,i]).all():
                count_w3 +=1
    print('count_w1:',count_w1) 
    print('count_w2:',count_w2) 
    print('count_w3:',count_w3) 
    if (count_w1>count_w2) and (count_w1>count_w3):
        print('This point belongs to class w1')
    elif (count_w2>count_w1) and (count_w2>count_w3):
        print('This point belongs to class w2')
    elif (count_w3>count_w1) and (count_w3>count_w2):
        print('This point belongs to class w3')
    elif (count_w3 == count_w1) or (count_w3 == count_w2) or (count_w2==count_w1):
        print('This point belongs to class either of two classes')
    else:
        print('This point belongs to class any of three classes')
    return
    
```
