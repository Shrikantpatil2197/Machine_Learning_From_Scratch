### Function for calculating P(given point) using parzon window:Parzon Window estimate

**ϕ((x − xi)/h) ∝ Exp[−(x − xi)^t (x − xi)/(2h^2)]**

```python
def Parzon(point,X,h):
    k = 0
    for i in range(len(X)):
        k = k + np.exp(-np.dot((point-X[:,i]).T,(point-X[:,i]))/(2*(h**2)))
    px_w = k/(len(X)*(h**3))
    return px_w
```

### a) Write a program to classify an arbitrary test point x based on the Parzen window estimates.
**Train your classifier using the three-dimensional data from three categories in the table above.
Set h = 1 and classify the following three points: (0.50, 1.0, 0.0)t, (0.31, 1.51, −0.50)t and (−0.3, 0.44, −0.1)t.**

```python
h = 1
```

**Define function to find class of given point**

```python
def Parson_classifier(point,X_W1,X_W2,X_W3,X_W,h):
    p_of_point_given_class_w1 = Parzon(point,X_W1,h)
     #p(point1|w1)

    p_of_point_given_class_w2 = Parzon(point,X_W2,h)
     #p(point1|w2)

    p_of_point_given_class_w3 = Parzon(point,X_W3,h)
     #p(point1|w3)

    total_p_point = Parzon(point,X_W,h) # X_W =all training examples
    #p(point1)

    p_of_class_w1_given_point = p_of_point_given_class_w1/total_p_point
    ##p(w1|point1) = p(point1|w1)/p(point1)
    print("p(class_w1|given_point):",p_of_class_w1_given_point)
    p_of_class_w2_given_point = p_of_point_given_class_w2/total_p_point
    ##p(w2|point1)
    print("p(class_w2|given_point):",p_of_class_w2_given_point)
    p_of_class_w3_given_point = p_of_point1_given_class_w3/total_p_point
    ##p(w3|point1)
    print("p(class_w3|given_point):",p_of_class_w3_given_point)

    if(p_of_class_w1_given_point>p_of_class_w2_given_point and p_of_class_w1_given_point>p_of_class_w3_given_point):
        print('given point belongs to class w1')
    elif (p_of_class_w2_given_point>p_of_class_w1_given_point and p_of_class_w2_given_point>p_of_class_w3_given_point):
        print('given point belongs to class w2')
    else:
        print('given point belongs to class w3')
    
    return
```

**1. point1 = (0.50, 1.0, 0.0) h=1**

```python
point1 = np.array([0.5,1,0])
Parson_classifier(point1,X_W1,X_W2,X_W3,X_W,h)
```

**2. point2 = (0.31, 1.51, − 0.50) h=1**

```python
point2 = np.array([0.31, 1.51, -0.5])
Parson_classifier(point2,X_W1,X_W2,X_W3,X_W,h)
```

**3. point3 = (− 0.3, 0.44, − 0.1) h=1**

```python
point3 = np.array([-0.3, 0.44, -0.1])
Parson_classifier(point3,X_W1,X_W2,X_W3,X_W,h)
```
