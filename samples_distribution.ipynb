```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

#### Q1_a)  Write a program to generate points according to a uniform distribution in a unit cube, âˆ’1/2 â‰¤ xi â‰¤ 1/2 for i = 1, 2, 3. Generate 10^4 such points.

```python
np.random.seed(42)  #so that results will be reproducible
x1 = np.random.uniform(-0.5,0.5,10**4)
x2 = np.random.uniform(-0.5,0.5,10**4)
x3 = np.random.uniform(-0.5,0.5,10**4)
```

```python
print('Uniform distribution -0.5 < (x1,x2,x3) < 0.5:ðŸ‘‡')
fig, ax = plt.subplots(1,3,figsize=(10,5))
ax[0].hist(x1)
ax[1].hist(x2)
ax[2].hist(x3)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x1,x2,x3,s=50,alpha=0.9);
```

#### Q1_b) Write a program to estimate the density at the origin    based on the above 10^4 points as a function of the size of a cubical window function of size h. Plot the estimate as a function of h, for 0 < h â‰¤ 1.

```python
np.random.seed(42)
h = np.random.uniform(0,1,size=1000) # 100 h values in between 0 and 1
counts = [] # list of no. of points in cube for different h values. 
for h_value in h:
    count = 0
    for i in range(10000):
        if (abs(x1[i])<h_value and abs(x2[i])<h_value and abs(x3[i])<h_value):
            count = count + 1
    counts.append(count)
```

```python
plt.scatter(h,counts)
plt.title('Density at the origin vs size of cubicle window(h)')
plt.xlabel('Size of cubicle window(h)')
plt.ylabel('Density at the origin')
plt.axvline(0.5,color='r',linestyle='--')
plt.xticks(np.arange(0,1,0.1))
plt.show()
```

#### Q1_c) Evaluate the density at the origin using n of your points and the volume of a cube window which just encloses n points. Plot your estimate as a function of n = 1  to 10^4


```python
plt.scatter(counts,h**3)
plt.title('volume of cubicle window(h**3) vs Density at the origin')
plt.ylabel('volume of cubicle window (h**3)')
plt.xlabel('Density at the origin')
plt.show()
```

#### Q1_d) Write a program to generate 10^4 points from a spherical Gaussian density (with Î£ = I) centered on the origin. Repeat (b) & (c) with the Gaussian data.

```python
# Let vector X = [x1,x2,x3] 
np.random.seed(42)
mean = np.zeros(3)
var = 1
cov = (var**2) * np.eye(3)
X = np.random.multivariate_normal(mean,cov,size=10000)
    
```

```python
print('Spherical Gaussian distribution mean=0, cov=1:ðŸ‘‡')
fig, ax = plt.subplots(1,3,figsize=(10,5))
ax[0].hist(X[:,0])
ax[1].hist(X[:,0])
ax[2].hist(X[:,0])
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0],X[:,1],X[:,2],s=50,alpha=0.9);
```

```python
x1 = X[:,0]
x2 = X[:,1]
x3 = X[:,2]
np.random.seed(42)
h = np.random.uniform(0,4,size=1000) # 100 h values in between 0 and 4
counts = [] # list of no. of points in cube for different h values. 
for h_value in h:
    count = 0
    for i in range(10000):
        if (abs(x1[i])<h_value and abs(x2[i])<h_value and abs(x3[i])<h_value):
            count = count + 1
    counts.append(count)
```

```python
plt.scatter(h,counts)
plt.title('Density at the origin vs size of cubicle window(h)')
plt.xlabel('Size of cubicle window(h)')
plt.ylabel('Density at the origin')
plt.xticks(np.arange(0,4,0.5))
plt.show()
```

```python
plt.scatter(counts,h**3)
plt.title('volume of cubicle window(h**3) vs Density at the origin')
plt.ylabel('volume of cubicle window (h**3)')
plt.xlabel('Density at the origin')
plt.show()
```
